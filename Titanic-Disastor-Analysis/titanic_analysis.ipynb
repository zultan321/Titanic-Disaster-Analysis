```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = sns.load_dataset('titanic')

print(df.head())

print('-' * 20)
print('SECTION 1: Exeutive summary')
print('-' * 20)
total_passengers = len(df)
survivors= df['survived'].sum()
highest_fare = df['fare'].max()
avg_passengers = df['age'].mean()
oldest_person = df['age'].max()
youngest_person = df['age'].min()

print('Passengers info ')
print('-'* 20)
print(f'{total_passengers} total passengers')
print(f'{survivors} passenger was survived')
print(f'${highest_fare:.0f} was the heighst paid fare (a fortune in 1912!)')
print()

print('Average age of passengers?')
print(f'{avg_passengers:.1f} years old by average  (most of the passengers was young adults)')
print(f"{oldest_person:.0f} is the oldest person's age")
print(f"{youngest_person} is the youngest person's age")

print()
print('-' * 20)
print('SECTION 2: Survived by gender')
print('-' * 20)
print('male vs female')

man = df[df['sex'] =='male'] ['survived'].sum()
women= df[df['sex'] =='female'] ['survived'].sum()

man_percent = df[df['sex'] == 'male'] ['survived'].mean() *100
women_percent = df[df['sex'] == 'female'] ['survived'].mean() *100

print(f'•{man} male with {man_percent:.0f}% of male passengers')
print(f'•{women} female with {women_percent:.0f}% of female passengers')

print()
print('-' * 20)
print('SECTION 3: Class survived insight')
print('-' * 20)
print('survived by class')

f_class = df[df['class'] == 'First'] ['survived'].sum()
s_class = df[df['class'] == 'Second'] ['survived'].sum()
th_class = df[df['class'] == 'Third'] ['survived'].sum()

print(f'First class:      {f_class} was survived')
print(f'Second class: {s_class} was survived')
print(f'Third class:     {th_class} was survived')

print()
print('-' *20)
print('SECTION 4: insight')
print('-' * 20)
print('Most surviveres are women and children')
print('First class have higher survived rate compaired to the other two classes ')
print(f'Someone paid a ${highest_fare:.0f}for fare')
print('Most passenger are youth(middle age) ')

print()
print('-' * 20)
print('SECTION 5: Visualitation')
print('-' * 20)
print('Survived by gender chart')

survived_by_gender = df[df['survived'] ==1]['sex'].value_counts()
plt.figure(figsize=(6,6))
plt.pie(
	survived_by_gender, 
	labels=survived_by_gender.index,
	autopct='%1.1f%%',
	colors=['gold', 'silver'],
	startangle=90
	)
plt.title('survived by gender rate ')
plt.savefig('survived_by_gender.png', dpi=300, bbox_inches='tight')
plt.show()

print()
print('-' * 20)
print('Survived by class chart')

survived_by_class = df[df['survived'] ==1] ['class'].value_counts()

plt.figure(figsize=(10,6))
plt.bar(survived_by_class.index, survived_by_class.values, color= [ 'gold', 'silver', 'skyblue'] )
plt.title('Survived by class')
plt.ylabel('Number of survivers')
plt.savefig('survived_by_gender.png', dpi=300, bbox_inches='tight')
plt.tight_layout()
plt.show()

print()
print('-' * 20)
print('Embark towns by chart')

town_embark_chart = df [ 'embark_town'].value_counts()
plt.figure(figsize=(10,6))
plt.bar(town_embark_chart.index, town_embark_chart.values, color = ['green', 'blue', 'yellow'])
plt.title('Char by embark town')
plt.xlabel('Embark town')
plt.ylabel('Number of passengers')
plt.savefig('survived_by_gender.png', dpi=300, bbox_inches='tight')

plt.tight_layout()
plt.show()
print("✅ REPORT GENERATION COMPLETE!")
print("=" * 40)
```
